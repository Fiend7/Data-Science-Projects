```python

data = data.drop(['Id'], axis = 1)
```

```python

x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# Start of Regression

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
## It transforms the data in such a manner that it has mean as 0 and standard deviation as
from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python

import graphviz
from sklearn import tree
classname =  ['setosa', 'versicolor', 'virginica']
dot_data = tree.export_graphviz(classifier, feature_names=data.columns[0:4], out_file=None,filled=True,class_names=classname,rounded=True,special_characters=True)
graph = graphviz.Source(dot_data, format="png") 
graph
```
